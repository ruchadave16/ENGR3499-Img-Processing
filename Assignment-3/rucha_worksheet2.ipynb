##### Question 1

##### Question 2
Median Filter

```python
array = np.random.rand(3, 3)
flattened = array.flatten()
print(f"Original array: \n{array}")

sorted_array = np.sort(flattened)
median = np.median(sorted_array)
print(f"\nThe median value of this array: {median}")

sorted_array = sorted_array.reshape(3, 3)
print(f"\nSorted array: \n{sorted_array}")
```

##### Question 3
Playing with different noise

```python
import skimage as ski

img = cv.imread('book.jpg')
noise1 = ski.util.random_noise(img, mode="s&p")
noise2 = ski.util.random_noise(img, mode='gaussian')
noise3 = ski.util.random_noise(img, mode="poisson")

fig, (ax1, ax2) = plt.subplots(2, 2)
fig.set_size_inches(10, 8)

ax1[0].imshow(img)
ax1[1].imshow(noise1)
ax2[0].imshow(noise2)
ax2[1].imshow(noise3)

ax1[0].title.set_text("Original Image")
ax1[1].title.set_text("Salt and Pepper Noise Image")
ax2[0].title.set_text("Gaussian Noise Image")
ax2[1].title.set_text("Poisson Noise Image")
```
